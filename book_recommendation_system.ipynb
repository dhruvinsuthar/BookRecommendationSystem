# POPULARITY BASED RECOMMENDATION SYSTEM

```python
ratings_with_name = ratings.merge(books,on="ISBN")
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num-ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean(['Book-Rating']).reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg-ratings'},inplace=True)
avg_rating_df
```

```python
popular_books = num_rating_df.merge(avg_rating_df,on='Book-Title').drop(columns=['User-ID'])
popular_books
```

```python
popular_books = popular_books[popular_books['num-ratings']>=400].sort_values('avg-ratings',ascending=False).head(20)
```

```python
popular_books
```

```python
popular_books = popular_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','num-ratings','avg-ratings']]
```

# COLLABORATIVE FILTERING BASED RECOMMENDATION SYSTEM

```python
a = ratings_with_name.groupby('User-ID').count()['Book-Rating']>100
rated_users = a[a].index
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(rated_users)]
filtered_ratings
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=20
famous_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    simialar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1], reverse=True)[1:11]

    data=[]
    for i in simialar_items:
        item=[]
        print(pt.index[i[0]])
        temp_df = books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    return data            
```

```python
test = final_ratings.drop_duplicates('Book-Title')[['ISBN','Book-Rating','Book-Title','Book-Author','Year-Of-Publication','Publisher','Image-URL-M']]
test
```

```python
recommend('Manhattan Hunt Club')
```
